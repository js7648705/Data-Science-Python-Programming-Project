## Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Plot the 30-day mortality rates for heart attack, heart failure and pneumonia

```python
file = pd.read_csv('outcome-of-care-measures.csv') # loading csv file
list_csv_col = [file.columns[1], file.columns[6],file.columns[10], file.columns[16],
                file.columns[22]]
# all columns are character columns-data values are in string format
df = pd.DataFrame(data = file[list_csv_col]) # converting list into dataframe
df.columns = ['Hospital.Name','State','Heart.attack', 'Heart.failure ','Pneumonia'] # naming 
# columns 2-5 columns are outcomes
```

```python
# Ploting histograms
plt.figure(figsize=(12,4))
df['Heart.attack']= pd.to_numeric(df['Heart.attack'],errors='coerce')
sns.histplot(df ['Heart.attack'],kde=False,bins=40,color = 'red')
plt.show()

plt.figure(figsize=(12,4))
df['Heart.faliure']= pd.to_numeric(df['Heart.failure '],errors='coerce')
sns.histplot(df ['Heart.faliure'],kde=False,bins=40)
plt.show()

plt.figure(figsize=(12,4))
df['Pneumonia']= pd.to_numeric(df['Pneumonia'],errors='coerce')
sns.histplot(df ['Pneumonia'],kde=False,bins=40,color = 'green')
plt.show()
```
